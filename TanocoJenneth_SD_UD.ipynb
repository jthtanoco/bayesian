```python
#SupervisedData
import pandas as pd
from sklearn.model_selection import train_test_split #The dataset was split into training (70%) ad testing (30%) sets using train_test_split.
from sklearn.linear_model import LogisticRegression #The alogorithm used in this dataset was Logistic Regression, a supervised learning algorithm suitable for binary classification problems. Unlike Linear Regression that predicts continuous values, Logistic Regression predicts categorical outcomes such as Pass or Fail.
from sklearn.metrics import accuracy_score #To measure how well the model predicts correctly, I used accuracy_score metrics.

url = "https://raw.githubusercontent.com/jthtanoco/bayesian/main/TanocoJenneth_SD-UD.xlsx"
supervised = pd.read_excel(url, sheet_name="SD_Locked")
#The dataset used was loaded from the Excel file/URL ("SD_Locked" sheet)

X = supervised[['Study Hours', 'Attendance (%)', 'Previous Score']]
y = supervised['Final Result (Label)']
#In this dataset, Study Hours, Attendance (%), Previous Score are treated as the input features (X), while Final Result (Label) as the target (y).

y = y.map({'Pass': 1, 'Fail': 0})
#The target column contained categorical values such as "Pass" or "Fail," which was mapped into numeric values (Pass = 1, Fail = 0) so the model could process them.

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train) #This trains the model using the training data that learns the relationship between input features and target.
predictions = model.predict(X_test) #Prediction model was used to predict outcomes on the test set.
#The model here was trained on the training set model.fit(X_train, y_train), then tested on unseen data model.predict(X_test).

print("Supervised Predictions:", predictions) #This displays the predicted results (Pass = 1, Fail = 0)
print("Accuracy:", accuracy_score(y_test, predictions)) #This calculates and prints the accuracy score of howm= many predictions were corrected compared to actual results.
```

```python
#UnsupervisedData
import pandas as pd
from sklearn.cluster import KMeans

url = "https://raw.githubusercontent.com/jthtanoco/bayesian/main/TanocoJenneth_SD-UD.xlsx"
unsupervised = pd.read_excel(url, sheet_name="UD_Locked")

X = unsupervised[['Customer Age', 'Annual Income', 'Spending Score']]

#usekMeansClustering
kmeans = KMeans(n_clusters=3, random_state=42)
unsupervised['Cluster'] = kmeans.fit_predict(X)

print(unsupervised.head())
```
